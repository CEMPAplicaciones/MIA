### Estad√≠stica descriptiva b√°sica

M√©todos estad√≠sticos b√°sicos que se pueden aplicar a arrays NumPy, esenciales para el an√°lisis exploratorio de datos en ciencia de datos: calcular el valor m√°ximo, m√≠nimo, media, desviaci√≥n est√°ndar y percentiles de un array.

```python
console.rule("[cyan]M√©todos varios a ejecutar sobre arrays[/cyan]")

# An√°lisis estad√≠stico completo de un array m√©dico
y = z
print("Y:", y)

print("\nValor m√°ximo:")
print("Max:", np.max(y))

print("\nValor m√≠nimo:")
print("Min:", np.min(y))

print("\nMedia:")
# La media es fundamental en an√°lisis cl√≠nicos (valores de referencia)
print("Media:", np.mean(y))

print("\nDesviaci√≥n est√°ndar:")
# La desviaci√≥n est√°ndar indica la variabilidad de los datos
print("Std:", np.std(y))

print("\nPercentil 25:")
# Los percentiles son cruciales para establecer rangos normales
print("Pct:", np.percentile(y, q=25))
print("Pct Q99:", np.percentile(y, q=99))

# Complemento de estad√≠stica descriptiva m√©dica
print("\nMediana:")
print("Mediana:", np.median(y))

print("\nVarianza:")
print("Var:", np.var(y))

print("\nRango (diferencia entre m√°ximo y m√≠nimo):")
# El rango es importante para evaluar la dispersi√≥n de valores cl√≠nicos
print("Rango:", np.ptp(y))

# ü©∫ EJERCICIOS: An√°lisis estad√≠stico
# Descomenta y practica con datos m√©dicos:

# Ejemplo 1: Analizar presiones arteriales
# presiones = np.array([120, 130, 110, 140, 125, 135, 115, 145])
# print(f"Presi√≥n media: {np.mean(presiones):.1f} mmHg")
# print(f"Presi√≥n m√°xima: {np.max(presiones)} mmHg")
# print(f"Percentil 95: {np.percentile(presiones, 95):.1f} mmHg")

# Ejemplo 2: Evaluar niveles de glucosa
# glucosa = np.array([85, 92, 105, 88, 110, 95, 98, 102])
# print(f"Glucosa - Media: {np.mean(glucosa):.1f}, Std: {np.std(glucosa):.1f}")
```

```python
# Funci√≥n auxiliar

def genera_datos_medicos():
  # Generar datos m√©dicos sint√©ticos para el ejemplo
  np.random.seed(42)  # Para reproducibilidad
  n_pacientes = 200

  # Crear datos m√©dicos realistas con correlaciones
  edad = np.random.normal(55, 15, n_pacientes)
  edad = np.clip(edad, 18, 90).astype(int)

  # Presi√≥n sist√≥lica correlacionada con edad (r‚âà0.4)
  presion_sistolica = 90 + 0.8 * edad + np.random.normal(0, 15, n_pacientes)
  presion_sistolica = np.clip(presion_sistolica, 80, 200).astype(int)

  # Crear diagn√≥sticos basados en presi√≥n
  diagnostico = []
  for presion in presion_sistolica:
    if presion < 120:
        diagnostico.append('Normal')
    elif presion < 140:
        diagnostico.append('Prehipertension')
    else:
        diagnostico.append('Hipertension')

  # Crear DataFrame m√©dico
  df_medico = pd.DataFrame({
    'edad': edad,
    'presion_sistolica': presion_sistolica,
    'peso': np.random.normal(70, 12, n_pacientes),
    'altura': np.random.normal(1.70, 0.10, n_pacientes),
    'diagnostico': diagnostico
  })
```

## Ejemplos adicionales

```python
console.rule("[cyan]Creaci√≥n de Arrays M√©dicos[/cyan]")

# 1. Datos de constantes vitales
print("ü©∫ SIMULACI√ìN: CONSTANTES VITALES")
print("-" * 40)

# Presi√≥n arterial sist√≥lica normal (rango 90-140 mmHg)
presion_sistolica = np.random.normal(120, 15, 100)  # media=120, std=15, n=100
presion_sistolica = np.clip(presion_sistolica, 80, 180)  # Limitar valores extremos

print(f"Presi√≥n sist√≥lica (primeros 10 pacientes): {presion_sistolica[:10].round(1)}")
print(f"Rango: {presion_sistolica.min():.1f} - {presion_sistolica.max():.1f} mmHg")

# 2. Frecuencia card√≠aca (rango normal 60-100 bpm)
frecuencia_cardiaca = np.random.normal(75, 10, 100)
frecuencia_cardiaca = np.clip(frecuencia_cardiaca, 50, 120).astype(int)

print(f"Frecuencia card√≠aca (primeros 10): {frecuencia_cardiaca[:10]}")

# 3. Temperatura corporal (rango normal 36-37.5¬∞C)
temperatura = np.random.normal(36.5, 0.5, 100)
temperatura = np.clip(temperatura, 35.0, 40.0)

print(f"Temperatura (primeros 10): {temperatura[:10].round(2)}")

# 4. Array multidimensional: Matriz de signos vitales
signos_vitales = np.column_stack((presion_sistolica, frecuencia_cardiaca, temperatura))
print(f"\nMatriz de signos vitales - Shape: {signos_vitales.shape}")
print("Columnas: [Presi√≥n Sist√≥lica, Frecuencia Card√≠aca, Temperatura]")
print(f"Primeras 5 filas:\n{signos_vitales[:5].round(2)}")
```

```python
console.rule("[cyan]An√°lisis Estad√≠stico de Constantes Vitales[/cyan]")

# An√°lisis detallado de cada par√°metro vital
parametros = ['Presi√≥n Sist√≥lica (mmHg)', 'Frecuencia Card√≠aca (bpm)', 'Temperatura (¬∞C)']
valores_normales = [(90, 140), (60, 100), (36.0, 37.5)]

print("üìä AN√ÅLISIS ESTAD√çSTICO DE CONSTANTES VITALES")
print("=" * 60)

for i, (parametro, rango_normal) in enumerate(zip(parametros, valores_normales)):
    datos = signos_vitales[:, i]

    print(f"\nüî¨ {parametro}")
    print(f"   üìà Media: {np.mean(datos):.2f}")
    print(f"   üìä Mediana: {np.median(datos):.2f}")
    print(f"   üìê Desviaci√≥n est√°ndar: {np.std(datos):.2f}")
    print(f"   üìè Rango: {np.min(datos):.2f} - {np.max(datos):.2f}")
    print(f"   üì¶ Cuartiles (Q1, Q3): {np.percentile(datos, [25, 75]).round(2)}")

    # Clasificaci√≥n cl√≠nica
    valores_normales_count = np.sum((datos >= rango_normal[0]) & (datos <= rango_normal[1]))
    porcentaje_normal = (valores_normales_count / len(datos)) * 100

    print(f"   ‚úÖ Valores normales: {valores_normales_count}/100 ({porcentaje_normal:.1f}%)")

    if porcentaje_normal >= 80:
        print(f"   üü¢ Estado: Poblaci√≥n saludable")
    elif porcentaje_normal >= 60:
        print(f"   üü° Estado: Revisar casos at√≠picos")
    else:
        print(f"   üî¥ Estado: Alta incidencia de valores an√≥rmales")

# Detecci√≥n de valores at√≠picos (outliers)
print(f"\nüö® DETECCI√ìN DE VALORES AT√çPICOS")
print("-" * 40)

for i, parametro in enumerate(parametros):
    datos = signos_vitales[:, i]
    Q1 = np.percentile(datos, 25)
    Q3 = np.percentile(datos, 75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = np.where((datos < limite_inferior) | (datos > limite_superior))[0]

    print(f"{parametro}: {len(outliers)} valores at√≠picos detectados")
    if len(outliers) > 0:
        print(f"   Pacientes: {outliers[:5].tolist()}{'...' if len(outliers) > 5 else ''}")
```

## Aplicai√≥n a imagen 2D: filtros (con scikit-image y matplotlib)

```python
console.rule("[cyan]Filtros sobre arrays (datos de imagen)[/cyan]")

import matplotlib.pyplot as plt
import numpy as np
from skimage import data, filters

# Cargar la imagen de muestra
image = data.camera()
print(f"Tama√±o de la imagen (array): {image.shape}")

# Crear una figura con subplots
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Mostrar la imagen original
ax[0].imshow(image)
ax[0].set_title('Imagen original')

# Aplicar filtro de desenfoque Gaussiano
blurred = filters.gaussian(image, sigma=16)
ax[1].imshow(blurred, cmap='gray')
ax[1].set_title('Filtro de desenfoque Gaussiano')

# Aplicar filtro Sobel para detectar bordes
edges = filters.sobel(image)
ax[2].imshow(edges, cmap='gray')
ax[2].set_title('Filtro Sobel para detecci√≥n de bordes')

# Ajustar los espacios entre subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, filters

# Cargar la imagen de muestra
image = data.camera()

# Calcular los percentiles 5 y 95 de la imagen
percentile_50 = np.percentile(image, 50)
percentile_99 = np.percentile(image, 99)

# Aplicar umbral a la imagen para eliminar valores por debajo del percentil 5
# y por encima del percentil 95
thresholded = np.clip(image, percentile_50, percentile_99)

# Crear una figura con subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Mostrar la imagen original
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Imagen original')

# Mostrar la imagen con valores fuera del rango recortados
ax[1].imshow(thresholded, cmap='gray')
ax[1].set_title('Valores recortados (5%-95%)')

plt.show()

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].hist(image)
ax[1].hist(thresholded)
```

```python
console.rule("[cyan]Pandas para Datos Sanitarios[/cyan]")

import pandas as pd
from tabulate import tabulate

print("üìã PANDAS: Gesti√≥n de Datasets M√©dicos")
print("=" * 50)

# Configurar opciones de visualizaci√≥n
pd.set_option('display.max_columns', 8)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 1000)


np.random.seed(42)  # Para reproducibilidad

# Crear dataset sint√©tico de pacientes
n_pacientes = 1000

# Generar datos demogr√°ficos
edades = np.random.normal(55, 20, n_pacientes).astype(int)
edades = np.clip(edades, 18, 95)

sexos = np.random.choice(['M', 'F'], n_pacientes, p=[0.52, 0.48])

# Generar datos cl√≠nicos realistas
presion_sistolica = np.random.normal(130, 20, n_pacientes)
presion_sistolica = np.clip(presion_sistolica, 90, 200)

# Colesterol total (mg/dL)
colesterol = np.random.normal(200, 40, n_pacientes)
colesterol = np.clip(colesterol, 120, 350)

# Glucosa en ayunas (mg/dL)
glucosa = np.random.normal(95, 15, n_pacientes)
glucosa = np.clip(glucosa, 70, 200)

# IMC (√çndice de Masa Corporal)
imc = np.random.normal(25, 5, n_pacientes)
imc = np.clip(imc, 16, 45)

# Diagn√≥sticos (simulados)
diagnosticos = np.random.choice([
    'Hipertensi√≥n', 'Diabetes', 'Dislipidemia', 'Obesidad', 'Normal'
], n_pacientes, p=[0.25, 0.15, 0.20, 0.15, 0.25])

# Crear DataFrame
datos_pacientes = pd.DataFrame({
    'ID_Paciente': [f'P{i:04d}' for i in range(1, n_pacientes + 1)],
    'Edad': edades,
    'Sexo': sexos,
    'Presion_Sistolica': presion_sistolica.round(0).astype(int),
    'Colesterol_Total': colesterol.round(0).astype(int),
    'Glucosa_Ayunas': glucosa.round(0).astype(int),
    'IMC': imc.round(1),
    'Diagnostico_Principal': diagnosticos
})

print("üìä DATASET M√âDICO GENERADO")
print("-" * 30)
print(f"Total de pacientes: {len(datos_pacientes):,}")
print(f"Variables registradas: {len(datos_pacientes.columns)}")

# Mostrar primeras filas
console.print("\n[bold]Registros de pacientes:[/bold]")
print(datos_pacientes)

# Informaci√≥n general del dataset
print(f"\nüìã INFORMACI√ìN DEL DATASET:")
print(f"   Dimensiones: {datos_pacientes.shape}")
print(f"   Memoria utilizada: {datos_pacientes.memory_usage(deep=True).sum() / 1024:.1f} KB")

# Tipos de datos
print(f"\nüî¢ TIPOS DE DATOS:")
for columna, tipo in datos_pacientes.dtypes.items():
    print(f"   {columna}: {tipo}")

```

```python
console.rule("[cyan]An√°lisis Epidemiol√≥gico[/cyan]")

print("ü¶† AN√ÅLISIS EPIDEMIOL√ìGICO DEL DATASET")
print("=" * 45)

# 1. An√°lisis demogr√°fico
print("üë• AN√ÅLISIS DEMOGR√ÅFICO:")
print("-" * 25)

# Distribuci√≥n por sexo
distribucion_sexo = datos_pacientes['Sexo'].value_counts()
print(f"Distribuci√≥n por sexo:")
for sexo, cuenta in distribucion_sexo.items():
    porcentaje = (cuenta / len(datos_pacientes)) * 100
    print(f"   {sexo}: {cuenta:,} pacientes ({porcentaje:.1f}%)")

# Estad√≠sticas de edad
print(f"\nEdad de la poblaci√≥n:")
print(f"   Media: {datos_pacientes['Edad'].mean():.1f} a√±os")
print(f"   Mediana: {datos_pacientes['Edad'].median():.1f} a√±os")
print(f"   Rango: {datos_pacientes['Edad'].min()} - {datos_pacientes['Edad'].max()} a√±os")

# Grupos etarios
bins_edad = [0, 30, 50, 65, 100]
labels_edad = ['< 30 a√±os', '30-50 a√±os', '50-65 a√±os', '> 65 a√±os']
datos_pacientes['Grupo_Etario'] = pd.cut(datos_pacientes['Edad'], bins=bins_edad, labels=labels_edad)

print(f"\nDistribuci√≥n por grupos etarios:")
for grupo, cuenta in datos_pacientes['Grupo_Etario'].value_counts().items():
    porcentaje = (cuenta / len(datos_pacientes)) * 100
    print(f"   {grupo}: {cuenta:,} pacientes ({porcentaje:.1f}%)")

# 2. Prevalencia de enfermedades
print(f"\nüè• PREVALENCIA DE ENFERMEDADES:")
print("-" * 35)

prevalencia = datos_pacientes['Diagnostico_Principal'].value_counts()
for enfermedad, casos in prevalencia.items():
    tasa = (casos / len(datos_pacientes)) * 100
    print(f"   {enfermedad}: {casos:,} casos ({tasa:.1f}%)")

# 3. An√°lisis de factores de riesgo cardiovascular
print(f"\n‚ù§Ô∏è FACTORES DE RIESGO CARDIOVASCULAR:")
print("-" * 40)

# Hipertensi√≥n (>140 mmHg)
hipertensos = (datos_pacientes['Presion_Sistolica'] >= 140).sum()
tasa_hipertension = (hipertensos / len(datos_pacientes)) * 100
print(f"   Hipertensi√≥n (‚â•140 mmHg): {hipertensos:,} casos ({tasa_hipertension:.1f}%)")

# Hipercolesterolemia (>240 mg/dL)
hipercolesterol = (datos_pacientes['Colesterol_Total'] >= 240).sum()
tasa_colesterol = (hipercolesterol / len(datos_pacientes)) * 100
print(f"   Hipercolesterolemia (‚â•240 mg/dL): {hipercolesterol:,} casos ({tasa_colesterol:.1f}%)")

# Diabetes (glucosa ‚â•126 mg/dL)
diabetes = (datos_pacientes['Glucosa_Ayunas'] >= 126).sum()
tasa_diabetes = (diabetes / len(datos_pacientes)) * 100
print(f"   Diabetes (‚â•126 mg/dL): {diabetes:,} casos ({tasa_diabetes:.1f}%)")

# Obesidad (IMC ‚â•30)
obesidad = (datos_pacientes['IMC'] >= 30).sum()
tasa_obesidad = (obesidad / len(datos_pacientes)) * 100
print(f"   Obesidad (IMC ‚â•30): {obesidad:,} casos ({tasa_obesidad:.1f}%)")

# 4. An√°lisis por grupos de riesgo
print(f"\nüéØ ESTRATIFICACI√ìN DE RIESGO:")
print("-" * 30)

# Pacientes con m√∫ltiples factores de riesgo
datos_pacientes['Riesgo_Multiple'] = (
    (datos_pacientes['Presion_Sistolica'] >= 140) &
    (datos_pacientes['Colesterol_Total'] >= 240) &
    (datos_pacientes['IMC'] >= 30)
)

alto_riesgo = datos_pacientes['Riesgo_Multiple'].sum()
tasa_alto_riesgo = (alto_riesgo / len(datos_pacientes)) * 100
print(f"   Pacientes con alto riesgo m√∫ltiple: {alto_riesgo:,} ({tasa_alto_riesgo:.1f}%)")
```
